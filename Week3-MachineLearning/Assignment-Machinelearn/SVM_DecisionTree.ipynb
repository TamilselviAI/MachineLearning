```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
datasets = pd.get_dummies(dataset, dtype=int,drop_first=True)
```

```python
datasets
```

```python
datasets.columns
```

```python
indep = datasets [['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dep = datasets [['Profit']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split (indep,dep,test_size=0.5, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
# Hyper Parameter inside function
regressor=DecisionTreeRegressor(criterion='poisson',splitter='random',max_features='log2')
regressor=regressor.fit(X_train,y_train)
```

```python
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
