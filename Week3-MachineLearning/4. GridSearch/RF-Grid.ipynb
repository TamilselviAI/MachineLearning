```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np
```

```python
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
datanomin = pd.get_dummies(dataset, dtype=int, drop_first = True)
```

```python
datanomin
```

```python
datanomin.columns
```

```python
independ = datanomin[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dep = datanomin[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independ,dep, test_size =0.30, random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion':['squared_error','friedman_mse','absolute_error','poisson'],'max_features':['sqrt','log2',None],'n_estimators':[50,100]}
```

```python
grid = GridSearchCV(RandomForestRegressor(), param_grid, refit =True, verbose=3,n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
re = grid.cv_results_
```

```python
print(re)
```

```python
grid.predictions = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,grid.predictions)
r_score
```

```python
print ( "R_score {} is : ", (grid.best_params_),r_score)
```
