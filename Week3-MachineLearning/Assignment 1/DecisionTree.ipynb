```python
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
datanomin = pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
datanomin
```

```python
datanomin.columns
```

```python
indep = datanomin[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dep = datanomin [['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split (indep,dep,test_size = 0.5, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor (criterion = 'poisson', max_features='log2',splitter='random')
regressor = regressor.fit(X_train,y_train)
```

```python
from sklearn import tree
tree.plot_tree(regressor)
plt
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
r_score
```
