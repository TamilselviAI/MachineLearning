```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset = pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
datanomin = pd.get_dummies(dataset,drop_first=True,dtype =int)
```

```python
datanomin
```

```python
datanomin.columns
```

```python
indep = datanomin [['User ID', 'Age', 'EstimatedSalary', 'Gender_Male']]
dep = datanomin [['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(indep,dep,test_size =0.5, random_state=0)
```

```python
from sklearn.svm import SVC
#classifier = SVC(kernel="sigmoid",C=1000)
```

```python

from sklearn.model_selection import GridSearchCV

param_grid = {'kernel':['linear','rbf','poly','sigmoid'],
             'gamma':['auto','scale'],
             'C':[10,100,1000,2000,3000]} 


grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
```

```python
grid.fit(X_train, y_train) 
re=grid.cv_results_
grid_predictions = grid.predict(X_test) 
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,grid_predictions)
```
